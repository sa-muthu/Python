# **Seaborn**

```python
###Class
import seaborn as sns                # For advanced statistical visualization
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set visual theme
sns.set_style("whitegrid")
```

```python
import seaborn as sns
print(sns.get_dataset_names())#get a list of all datasets that are in-built in Seaborn
```

```python
# Load built-in datasets
df = sns.load_dataset('car_crashes')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
help(iris)
# View first few rows
print(df.head())
print(df.tail())
```

```python
print(iris.head())
```

```python
# Basic statistical summary
print("SUMMARY STATISTICS ")
print(iris.describe())
# Check data types
print("\nDATA TYPES")
print(iris.dtypes)
# Select only numeric columns
numeric_iris = iris.select_dtypes(include='number')
# Correlation matrix
print("\n CORRELATION MATRIX ")
corr = numeric_iris.corr()
print(corr)
# Covariance matrix
print("\nCOVARIANCE MATRIX ")
cov = numeric_iris.cov()
print(cov)
```

Mean → central tendency

Std → variability

Correlation → strength of relationship (-1 to 1)

Covariance → direction of relationship

## **Heatmap**

```python
import seaborn as sns                # For advanced statistical visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
# Heatmap visualization
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f')

plt.title("Correlation Heatmap")
plt.show()
```

## **Scatterplot**

```python
sns.scatterplot(x="speeding", y="alcohol", data=df)
plt.xlabel("Speeding")
plt.ylabel("Alcohol")

plt.title("Speeding vs Alcohol")
plt.show()
```

Each point = one observation

Detects outliers

Shows linear/non-linear trends

Useful for hypothesis formation

```python
sns.regplot(x="speeding", y="alcohol", data=df)
plt.title("Regression Trend")
plt.show()
```

Line shows best-fit regression

Confidence interval indicates uncertainty

Helps in predictive modeling

## **KDEPLOT**

```python
sns.kdeplot(data=df, x="speeding", fill=True)
plt.title("Density Distribution")
plt.show()
```

Smooth probability density

Identifies skewness

Better than histogram for distribution

## Histogram

```python
sns.histplot(data=df, x="alcohol", kde=True, bins=20)
plt.title("Alcohol Distribution")
plt.show()
```

Histogram → frequency

KDE → smooth density

Helps detect skewed data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create jointplot
g = sns.jointplot(
    x="speeding",
    y="alcohol",
    data=df,
    kind="reg",
    height=7,
    ratio=4,
    color="purple",
    marginal_kws={"bins": 15, "fill": True}
)

# Add title properly (important for jointplot)
g.fig.suptitle("Relationship Between Speeding and Alcohol-Related Crashes", fontsize=14)

# Adjust layout (very important)
g.fig.subplots_adjust(top=0.92)

plt.show()
```

Combines scatter + histogram

Shows joint + marginal distributions

Useful for bivariate analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
print(tips.head())
# Add labels
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
print(tips['day'])
# Title
plt.title("Boxplot of Total Bill")
plt.show()
```

Middle line → median

Box → interquartile range (IQR)

Whiskers → spread

Points outside → outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18,5))

#  1. Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[0].set_title("Violin Plot")
axes[0].set_xlabel("Day")
axes[0].set_ylabel("Total Bill ($)")

#  2. Histogram
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[1])
axes[1].set_title("Histogram + KDE")
axes[1].set_xlabel("Total Bill ($)")
axes[1].set_ylabel("Frequency")

#  3. Bar Plot
sns.barplot(x="day", y="total_bill", data=tips, ax=axes[2])
axes[2].set_title("Average Bill by Day")
axes[2].set_xlabel("Day")
axes[2].set_ylabel("Average Total Bill ($)")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

Shows all pairwise relationships

Diagonal → distributions

Useful for feature separability

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
g = sns.pairplot(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

Customizable pair plot

More control than pairplot

Used in research visualization
